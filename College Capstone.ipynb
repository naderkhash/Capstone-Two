```python
import pandas as pd
# Opening each of the CSV files
degrees = pd.read_csv(r'C:\Users\Nader\Desktop\Coding Stuff\Springboard\Caapstone\College Idea\degrees-that-pay-back.csv')
salaries = pd.read_csv(r'C:\Users\Nader\Desktop\Coding Stuff\Springboard\Caapstone\College Idea\salaries-by-region.csv')
college = pd.read_csv(r'C:\Users\Nader\Desktop\Coding Stuff\Springboard\Caapstone\College Idea\salaries-by-college-type.csv')
tuition = pd.read_csv(r'C:\Users\Nader\Desktop\Coding Stuff\Springboard\Caapstone\College Idea\tuition_cost.csv')
# Cleaning up college data
college_clean = college.drop(columns=['Mid-Career 25th Percentile Salary', 'Mid-Career 10th Percentile Salary', 'Mid-Career 75th Percentile Salary', 'Mid-Career 90th Percentile Salary'])
college_clean.rename(columns = {'Starting Median Salary': 'Starting Median Salary (College)', 'Mid-Career Median Salary' : 'Mid-Career Median Salary (College)'})
college_clean
```

```python
# Dropping unneccessary columns from salaries
salaries_clean = salaries.drop(columns = ['Mid-Career 25th Percentile Salary', 'Mid-Career 10th Percentile Salary', 'Mid-Career 75th Percentile Salary', 'Mid-Career 90th Percentile Salary'])
salaries_clean
```

```python
# Merging the salaries and college datasets 
salaries_college = pd.merge(salaries_clean, college_clean, how='left', on='School Name')
salaries_college
Colleges = salaries_college.drop(columns=['Starting Median Salary_y', 'Mid-Career Median Salary_y'])
Colleges = Colleges.rename(columns = {'Starting Median Salary_x': 'Starting Median Salary', 'Mid-Career Median Salary_x': 'Mid Career Median Salary'})
```

```python
tuition.columns
tuition = tuition.rename(columns={'name': 'School Name'})
```

```python
# Cleaning the degrees dataset
degrees_clean = degrees.drop(columns=['Mid-Career 25th Percentile Salary', 'Percent change from Starting to Mid-Career Salary','Mid-Career 10th Percentile Salary', 'Mid-Career 75th Percentile Salary', 'Mid-Career 90th Percentile Salary'])
```

```python
tuition = tuition.drop(columns =['type', 'state', 'state_code', 'degree_length', 'in_state_total', 'room_and_board', 'out_of_state_total'])
```

```python
tuition = tuition.set_index('School Name')
```

```python
# Merging colleges and tuitions
colleges_tuition = pd.merge(Colleges, tuition, how='left', on='School Name')
```
